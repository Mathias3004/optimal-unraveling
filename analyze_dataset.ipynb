```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import sys
import progressbar
```

```python
folder = "store_XXZ_t_500_projUp/"

# load data
with open(folder + 'dat_S.pickle', 'rb') as handle:
    dat_S = pickle.load(handle)
    
# load variables used
v_N = dat_S["v_N"] 
v_gamma = dat_S["v_gamma"]
Jx = dat_S["Jx"]
Jz = dat_S["Jz"]
optims = dat_S["optims"]
v_it = dat_S["v_it"]
n_sample = dat_S["n_sample"]

!ls $folder
```

```python
# obtain mean over collected trajectories
dat_S_mean_traj = {}
for optim in optims:
    dat_S_mean_traj[optim] = []
    for i_N, N in enumerate(list(v_N)):
        dat_S_mean_traj[optim].append( np.mean(dat_S[optim][i_N], axis=-1) )
```

```python
# compute timelines
N = 20
#gamma_plot = [0.1, 0.5, 1.5, 3.]
gamma_plot = v_gamma[0:]
i_N = np.argmin(np.abs(N-v_N))
i_l = N // 2

# color code
clr = plt.cm.jet_r(gamma_plot/np.max(gamma_plot)) 

plt.figure(figsize=(16, 8), dpi=80)
for i_p, gamma in enumerate(list(gamma_plot)):
    i_gamma = np.argmin(np.abs(gamma-v_gamma))
    plt.plot(dat_S_mean_traj["none"][i_N][i_gamma,:,i_l], 
            label=r"bare $\gamma={:.2f}$".format(gamma), linestyle="-", color=clr[i_p])
    plt.plot(dat_S_mean_traj["local"][i_N][i_gamma,:,i_l], 
            label=r"opt $\gamma={:.2f}$".format(gamma), linestyle=":", color=clr[i_p])
    
plt.legend(loc="center left")
plt.xlabel("time")
plt.ylabel("Entropy")
plt.xlim(left=.0)
plt.ylim(bottom=.0)
```

```python
# clean plot with results or bare, on-site quantum jumps

# compute timelines
N = 20
#gamma_plot = [0.1, 0.5, 1.5, 3.]
gamma_plot = v_gamma[0:]
i_N = np.argmin(np.abs(N-v_N))
i_l = N // 2

# color code
clr = plt.cm.jet_r(gamma_plot/np.max(gamma_plot)) 

plt.figure(figsize=(8, 4.5), dpi=80)

for i_p, gamma in enumerate(list(gamma_plot)):
    i_gamma = np.argmin(np.abs(gamma-v_gamma))
    dat_plot = dat_S_mean_traj["none"][i_N][i_gamma,:,i_l]
    plt.plot([0.] + list(dat_plot[dat_plot > 1e-5]), 
            label=r"$\gamma={:.2f}$".format(gamma), linestyle="-", color=clr[i_p],linewidth=2)
    
#plt.legend(reversed(plt.legend().legendHandles), bbox_to_anchor=(1.05, .95))
plt.xlabel(r"time")
plt.ylabel("Half-chain entropy")
plt.xlim(left=.0, right=500.)
plt.ylim(bottom=.0)

labels = plt.legend(bbox_to_anchor=(1.05, .97))
plt.tight_layout()
#plt.savefig("figs/bare_dephasing.pdf")
```

```python
# clean plot with results or bare, on-site quantum jumps

# compute timelines
N = 20
#gamma_plot = [0.1, 0.5, 1.5, 3.]
gamma_plot = v_gamma[0:]
i_N = np.argmin(np.abs(N-v_N))
i_l = N // 2

# color code
clr = plt.cm.jet_r(gamma_plot/np.max(gamma_plot)) 

plt.figure(figsize=(8, 4.5), dpi=80)

dat_ref = dat_S_mean_traj["none"][i_N][0,100:,i_l]
bare_sat_S = np.mean(dat_ref[dat_ref>0.])
plt.hlines(bare_sat_S, 0, 500, color="k", linestyle=":")
plt.text(200, bare_sat_S + 0.1, "steady entanglement without optimization")

for i_p, gamma in enumerate(list(gamma_plot)):
    i_gamma = np.argmin(np.abs(gamma-v_gamma))
    dat_plot = dat_S_mean_traj["local"][i_N][i_gamma,:,i_l]
    plt.plot([0.] + list(dat_plot[dat_plot > 1e-5]), 
            label=r"$\gamma={:.2f}$".format(gamma), linestyle="-", color=clr[i_p], linewidth=2)
    
#plt.legend(reversed(plt.legend().legendHandles), bbox_to_anchor=(1.05, .95))
plt.xlabel(r"time")
plt.ylabel("Half-chain entropy")
plt.xlim(left=.0, right=500.)
plt.ylim(bottom=.0)

labels = plt.legend(bbox_to_anchor=(1.05, .97))
plt.tight_layout()
#plt.savefig("figs/local_optimized_dephasing.pdf")
```
