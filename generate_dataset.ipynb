```python
# folder to process for datgen
folder = "store_XXZ_t_500/"

# list of all parameter generated

# varying parameters
v_N = np.arange(8, 21, 4)
v_gamma = np.linspace(.2,3.,8)

# fixed params
Jx = -1.
Jz = 0.

# optimizations
optims = ["none", "local"]

# time points to collect
v_it = np.arange(1,501)

# number of points for each time points (number of lines in a file)
n_sample = 128

pre = "dat__"

# some things to save
N_N, N_gamma = v_N.size, v_gamma.size
N_it = v_it.size

# show folder content
!ls $folder
```

## collect all S files in folder and turn them into pickle

```python
# collect all entropies, store in np array

# entropies
zero_dat = [np.zeros((N_gamma, N_it, N+1, n_sample )) for N in v_N]
dat_S = {}
for optim in optims:
     dat_S[optim] = zero_dat

# <Sz> 
zero_dat = [np.zeros((N_gamma, N_it, N, n_sample )) for N in v_N]
dat_sz = {}
for optim in optims:
     dat_sz[optim] = zero_dat

df_missing = pd.DataFrame(columns = ['optim', 'N', 'gamma','it']) # collect files that are not found

# load all text files to collect for analysis

# the loop with progress bar
bar = progressbar.ProgressBar(maxval=len(optims)*N_N*N_gamma*N_it, \
widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])

count = 0
bar.start()  

dat_S = {}
for optim in optims:
    dat_S_N = []
    for i_N, N in enumerate(list(v_N)):
        dat_S_N_gamma = np.zeros((N_gamma, N_it, N+1, n_sample ))
        for i_gamma, gamma in enumerate(list(v_gamma)):
            for i_it, it in enumerate(list(v_it)):
                suffix = "opt_{}_N_{}_Jx_{:.2f}_Jz_{:.2f}_gamma_{:.2f}_it_{}".format(optim,N,Jx,Jz,gamma,it)
                fn = folder + pre + suffix + "_S.txt"
                try:
                    data = np.loadtxt(fn)
                    dat_S_N_gamma[i_gamma,i_it,:,:] = data.T
                except:
                    print("file " + fn + " does not exist")
                    df_missing = df_missing.append({"optim":optim, "N":N, "gamma":gamma, "it":it}, ignore_index=True)
                
                count += 1
                bar.update(count)
        dat_S_N.append(dat_S_N_gamma)
    dat_S[optim] = dat_S_N
bar.finish()
```

```python
# print values that have gone wrong in time series (good if it's empty)
df_missing[['optim', 'N', 'gamma']].drop_duplicates().sort_values("gamma")
```

```python
import pickle

# add imprtant variables to dict
dat_S["v_N"] = v_N
dat_S["v_gamma"] = v_gamma
dat_S["Jx"] = Jx
dat_S["Jz"] = Jz
dat_S["optims"] = optims
dat_S["v_it"] = v_it
dat_S["n_sample"] = n_sample

# dump data in pickle
with open(folder + '/dat_S.pickle', 'wb') as handle:
    pickle.dump(dat_S, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# obtain mean over collected trajectories
dat_S_mean_traj = {}
for optim in optims:
    dat_S_mean_traj[optim] = []
    for i_N, N in enumerate(list(v_N)):
        dat_S_mean_traj[optim].append( np.mean(dat_S[optim][i_N], axis=-1) )
```

```python
# compute timelines
N = 20
#gamma_plot = [0.1, 0.5, 1.5, 3.]
gamma_plot = v_gamma[0:]
i_N = np.argmin(np.abs(N-v_N))
i_l = N // 2

# color code
clr = plt.cm.jet_r(gamma_plot/np.max(gamma_plot)) 

plt.figure(figsize=(16, 8), dpi=80)
for i_p, gamma in enumerate(list(gamma_plot)):
    i_gamma = np.argmin(np.abs(gamma-v_gamma))
    plt.plot(dat_S_mean_traj["none"][i_N][i_gamma,:,i_l], 
            label=r"bare $\gamma={:.2f}$".format(gamma), linestyle="-", color=clr[i_p])
    plt.plot(dat_S_mean_traj["local"][i_N][i_gamma,:,i_l], 
            label=r"opt $\gamma={:.2f}$".format(gamma), linestyle=":", color=clr[i_p])
    
plt.legend(loc="center left")
plt.xlabel("time")
plt.ylabel("Entropy")
plt.xlim(left=.0)
plt.ylim(bottom=.0)
```
